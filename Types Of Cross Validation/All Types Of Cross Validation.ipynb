```python
import pandas as pd
df=pd.read_csv('cancer_dataset.csv')
df.head()
```

```python
###  Independent And dependent features
X=df.iloc[:,2:]
y=df.iloc[:,1]
```

```python
X.head()
```

```python
X=X.dropna(axis=1)
```

```python
X.head()
```

```python
y.value_counts()
```

### 1.HoldOut Validation Approach- Normal Train And Test Split

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=50)
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
result = model.score(X_test, y_test)
print(result)
```

### 2.K Fold Cross Validation


```python
from sklearn.model_selection import KFold
model=DecisionTreeClassifier()
kfold_validation=KFold(10)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
results=cross_val_score(model,X,y,cv=kfold_validation)
print(results,"\n")
print(np.mean(results))
```

### 3. Stratified K-fold Cross Validation
We use stratified K-fold to cope with class imbalances in the data set. 

Stratified K-fold maintains the class proportions by splitting the data set in such a way that they contain approximately the same proportions of labels as in the original data set.
This strategy guarantees that when the data set is unbalanced, one class of the data is not over-represented. Details on video.

```python
# Proportion is matter
from sklearn.model_selection import StratifiedKFold
skfold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
scores=cross_val_score(model,X,y,cv=skfold)
print(np.mean(scores))
```

```python
scores
```

### 4.Leave One Out Cross Validation(LOOCV)

```python
from sklearn.model_selection import LeaveOneOut
model=DecisionTreeClassifier()
leave_validation=LeaveOneOut()
results=cross_val_score(model,X,y,cv=leave_validation)
```

```python
results
```

```python
print(np.mean(results))
```

### 5. Repeated Random Test-Train Splits
This technique is a hybrid of traditional train-test splitting and the k-fold cross-validation method. In this technique, we create random splits of the data in the training-test set manner and then repeat the process of splitting and evaluating the algorithm multiple times, just like the cross-validation method.

```python
from sklearn.model_selection import ShuffleSplit
model=DecisionTreeClassifier()
ssplit=ShuffleSplit(n_splits=10,test_size=0.30)
results=cross_val_score(model,X,y,cv=ssplit)
```

```python
results
```

```python
np.mean(results)
```
