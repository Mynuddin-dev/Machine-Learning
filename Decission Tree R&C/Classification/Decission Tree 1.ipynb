#### Decision Tree Implementation on Cancer Dataset
- https://www.kaggle.com/nisasoylu/decision-tree-implementation-on-cancer-dataset

```python
data= pd.read_csv('breastCancer.csv')
```

```python
data
```

```python
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
```

```python
data
```

```python
data.diagnosis.unique()
```

```python
import numpy as np
```

```python
data.diagnosis = np.where(data.diagnosis == "M",1,0)
data
```

```python
import numpy as np
y=data.diagnosis.values
#normalization
x_data=data.drop(["diagnosis"], axis=1)
x=(x_data-np.min(x_data))/(np.max(x_data) - np.min(x_data))   # Normalization 
```

```python
y
```

```python
x
```

```python
data.diagnosis.unique()
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier()
dt.fit(x_train,y_train)
print("score :",dt.score(x_test,y_test))
```

```python
print('Training set score: {:.4f}'.format(dt.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(dt.score(x_test, y_test)))
```

#### Iris Dataset

```python
from sklearn.tree import export_graphviz
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df_iris = sns.load_dataset("iris")
df_iris
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
df_iris.columns
```

```python
df_iris_X = df_iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
df_iris_y = df_iris.species
```

```python
df_iris_X
```

```python
df_iris_y
```

```python
tree_clf = DecisionTreeClassifier()
tree_clf.fit(df_iris_X, df_iris_y)
```

```python
tree_clf.predict([[5.0 ,3.6,1.4,0.2]])
```

```python
df_iris_y.unique()
```

```python
text_representation = tree.export_text(tree_clf)
print(text_representation)
```

- https://mljar.com/blog/visualize-decision-tree/

```python
df_iris_X.columns
```

```python
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(tree_clf, 
                   feature_names=df_iris_X.columns,  
                   class_names=df_iris_y.unique(),
                   filled=True)
```

```python
# Visualize the trained Decision Tree by export_graphviz() method

from sklearn.tree import export_graphviz
from sklearn import tree
import graphviz
labels = df_iris_X.columns

dot_data = tree.export_graphviz(tree_clf, 
                  feature_names=labels,  
                  class_names=df_iris_y.unique(),  
                  filled=True, rounded=True,  
                  special_characters=True,
                   out_file=None,
                           )
graph = graphviz.Source(dot_data)
graph
```
