```python
x = df.drop(["Profit"] , axis = 1)
x
```

```python
y = df["Profit"]
y
```

```python
dummies = pd.get_dummies(df["Area"])
dummies
```

```python
x= pd.concat([x ,dummies] , axis = "columns")
x = x.drop(["Area"] , axis = "columns")
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x , y ,test_size = .25 , random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
regobj = LinearRegression()
```

```python
regobj.fit(xtrain , ytrain)
```

```python
prediction = regobj.predict(xtest)
prediction
```

```python
regobj.score(xtest , ytest)
```

<h2>R-Squared Value</h2>

```python
from sklearn.metrics import r2_score
```

```python
score = (r2_score(ytest , prediction))*100
score
```
